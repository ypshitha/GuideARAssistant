```python
#Bubble sort
def bubblesort(nums):
    n= len(nums)
    is_swap=False
    for i in range(n-2,-1,-1):
        for j in range(0,i+1):
            if nums[j]>nums[j+1]:
                nums[j],nums[j+1]=nums[j+1],nums[j]
                is_swap=True
        if is_swap==False:
            break
    return nums
nums = [5,7,8,4,1,6,9,2]
result=bubblesort(nums)
print(result)
print(result[::-1])
```

```python
#selection sort
def insertionsort(nums):
    n= len(nums)
    for i in range(1,n):
        j= i-1
        key= nums[i]
        while j >=0 and nums[j]>key:
            nums[j+1]=nums[j]
            j =j-1
        nums[j+1]=key
    return nums
nums=[5,7,8,4,1,6,9,2]
result=insertionsort(nums)
print(result)
print(result[::-1])
```

```python
#merge sort 
def mergearray(left,right):
    i=0
    j=0
    n=len(left)
    m=len(right)
    result=[]
    while i<n and j<m :
        if left[i]<=right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    if i<n:
        while i<n :
            result.append(left[i])
            i+=1
    if j<m:
        while j<m:
            result.append(right[j])
            j+=1
    return result
def mergesort(arr):
    if len(arr)<=1:
        return arr
    mid = len(arr)//2
    left = mergesort(arr[:mid])
    right=mergesort(arr[mid:])
    return mergearray(left,right)
nums = [5,7,8,4,1,6,9,2]
result= mergesort(nums)
print(result)
print(result[::-1])
```

```python
#quicksort
def partition(nums,low,high):
    i=low
    j=high
    pivot=nums[i]
    while i<j:
        while nums[i]<=pivot and i<=high-1:
            i=i+1
        while nums[j]>pivot and j>=low+1:
            j=j-1
        if i<j:
            nums[i],nums[j]=nums[j],nums[i]
    nums[low],nums[j]=nums[j],nums[low]
    return j
def quicksort(nums,low,high):
    if low<high:
        pindex=partition(nums,low,high)
        quicksort(nums,low,pindex-1)
        quicksort(nums,pindex+1,high)
nums=[5,7,8,4,1,6,9,2]
quicksort(nums,0,len(nums)-1)
print(nums)
print(nums[::-1])
```

```python
#right rotate k times
#method 1
def rotate(nums,r):
    n=len(nums)
    rotations=r%n
    for _ in range(0,rotations):
        e=nums.pop()
        nums.insert(0,e)
    return nums
nums=[1,2,3,4,5,6,7]
print(rotate(nums,4))
#method 2
def rotate(nums,k):
    n=len(nums)
    k=k%n
    nums[:]=nums[n-k:]+nums[:n-k]
    return nums
nums=[1,2,3,4,5,6,7]
print(rotate(nums,4))
#method 3
def reverse(nums,left,right):
    n=len(nums)
    while left<right:
        nums[left],nums[right]=nums[right],nums[left]
        left+=1
        right-=1
    return nums
def rotate(nums, k):
    n = len(nums)
    k = k % n   
    reverse(nums, n - k, n - 1)
    reverse(nums, 0, n - k - 1)
    reverse(nums, 0, n - 1)
nums = [1, 2, 3, 4, 5, 6, 7]
k = 4
rotate(nums, k)
print(nums)
        
```

```python
#moving the zeroes to end
#method 1(Brute force)
def func(nums):
    temp=[]
    n=len(nums)
    for i in range(0,n):
        if nums[i]!=0:
            temp.append(nums[i])
    nz=len(temp)
    for i in range(0,nz):
        nums[i]=temp[i]
    for i in range(nz,n):
        nums[i]=0
    return nums
nums=[1,0,2,0,3,0,5]
print(func(nums))
#method 2 (optimal)
def func(nums):
    n=len(nums)
    i=0
    if n==1:
        return nums
    while i<n:
        if nums[i]==0:
            break
        i+=1
    if i==n:
        return nums
    j=i+1
    while j<n :
        if nums[j]!=0:
            nums[i],nums[j]=nums[j],nums[i]
            i+=1
        j+=1
    return nums
nums=[1,0,2,0,3,0,5]
print(func(nums))
```

```python
#finding the missing number i the given array
#brute force method 
def missing1(nums):
    n=len(nums)
    for i in range(0,n+1):
        if i not in nums:
            return i
#better solution using dictionary
def missing2(nums):
    n=len(nums)
    freq={}
    for i in range(0,n+1):
        freq[i]=0
    for num in nums:
        freq[num]=1
    for k,u in freq.items():
        if u==0:
            return k
#optimal solution
def missing3(nums):
    n=len(nums)
    return ((n*(n+1))//2)-sum(nums)

nums=[0,1,2,3,4,5,6,7,9]
solution=missing3(nums)
print(solution)
```

```python
#finding the maximum number of consective ones
def findMaxConsecutiveOnes(nums):
    n=len(nums)
    count=0
    max_count=0
    for i in range(0,n):
        if nums[i]==1:
            count+=1
        else:
            max_count=max(max_count,count)
            count=0
    return max(max_count,count)
nums=[1,1,0,1,1,1,1,0]
solution=findMaxConsecutiveOnes(nums)
print(solution)

       
```

```python
#two sum
#using 2 pointers
def twosum(nums,target):
    n=len(nums)
    for i in range(0,n-1):
        for j in range(1,n):
            if nums[i]+nums[j]==target:
                return [i,j]
nums=[1,2,3,4,5,6,9]
target=13
solution=twosum(nums,target)
print(solution)
#using the dictionary
def twosums(nums,target):
    n=len(nums)
    hash_map={}
    for i in range(0,n):
        remaining=target-nums[i]
        if remaining in hash_map:
            return [hash_map[remaining],i]
        else:
            hash_map[nums[i]]=i
            nums=[1,2,3,4,5,6,9]
target=13
solution=twosums(nums,target)
print(solution)
```

```python
def maxsubarray(nums):
    n=len(nums)
    maximum=float("-inf")
    
    for i in range(0,n):
        total=0
        for j in range(i,n):
            total+=nums[j]
            if total>maximum:
                maximum=total
    return maximum
nums=[1,2,3,4,5,6,9]
solution=maxsubarray(nums)
print(solution)
#kadene solution
def maxsubarray1(nums):
    n=len(nums)
    maximum=float("-inf")
    total=0
    for i in range(0,n):
        total+=nums[i]
        maximum=max(total,maximum)
        if total<0:
            total=0
    return maximum
nums=[1,2,3,4,5,6,9]
solution=maxsubarray1(nums)
print(solution)
```
